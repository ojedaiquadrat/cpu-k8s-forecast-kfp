kserve to serve models uses defaul images that can picked, theses images should have the same version libraries than the one used to train and log the model in the trainign part in the pipeline. there are two approaches:

1. Create an image or conda env with the the same libraries than the kserve images

2. Create an customized image with the same libraries in the training phase to register that image as **Serving Runtime**

# Training Pipeline
#### For the The Sunrise&G server workspace is : /home/eduardo/Documents/Sunrise/Workspace/training-pipeline

Steps: 

1. Create a secret with the minio user and pass so the kserve can access the buckets

**CODE:**
```

# Giving acces to my kubeflow to access the artifacts in the mlflow-MinIO 
## In KServe v0.15 the S3/MiniO settings must be annotated on the ServiceAccount (not the Secret)
apiVersion: v1
kind: Secret
metadata:
  name: minio-mlflow-artifacts
  namespace: lstm-iqu
type: Opaque
stringData:
  AWS_ACCESS_KEY_ID: "admin"
  AWS_SECRET_ACCESS_KEY: "ySO5ISk7Eq"
```



2. Create a secret in case access to dockerhub to pull images is needed if the repos are private (no needed here)
**CODE**
```
docker crednetial inmy dockerhub
#kubectl create secret docker-registry dockerhub-pull-secret \
#  --docker-username=jhofydu \
#  --docker-password=Elnacional987! \
#  --docker-server=https://index.docker.io/v1/ \
#  --namespace=lstm-iqu

```


3. A Service Account is needed to give the permission to kserve to access Minio and docker in case needed. By default the kserve uses the service accoutn called **sa-private-mlflow**

**CODE**:
```

apiVersion: v1
kind: ServiceAccount
metadata:
  name: sa-private-mlflow
  namespace: lstm-iqu
  annotations:
    serving.kserve.io/s3-endpoint: sunrise-minio.mlflow.svc.cluster.local:5081
    serving.kserve.io/s3-usehttps: "0"
    serving.kserve.io/s3-verifyssl: "0"
    serving.kserve.io/s3-region: us-east-1
    serving.kserve.io/s3-useanoncredential: "false"
    serving.kserve.io/s3-usevirtualbucket: "false"
    serving.kserve.io/s3-credentials-secret-name: minio-mlflow-artifacts
    serving.kserve.io/storageSecretName: minio-mlflow-artifacts
#imagePullSecrets:
#  - name: dockerhub-pull-secret
secrets:
  - name: minio-mlflow-artifacts
```

4. Finally the Inference service can be created. in this case we use the default runtime **kserve-mlserver such image is (seldonio/mlserver:1.5.0)**. You can pull wioth docker to check the version libraries, python, pytorch,mlflow

```
apiVersion: "serving.kserve.io/v1beta1"
kind: "InferenceService"
metadata:
  name: "test-24-inference"
  namespace: lstm-iqu 
spec:
  predictor:
    serviceAccountName: sa-private-mlflow
    model:
      modelFormat:
        name: mlflow
      #protocolVersion: v2
      storageUri: "s3://mlflow/27/5a0bfc55915a4510a2de21fa842d64aa/artifacts/model"
      runtime: kserve-mlserver

```
